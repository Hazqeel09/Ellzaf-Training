```python
import requests

API_KEY = 'get your key here: https://openweathermap.org/price'
CITY_NAME1 = 'Kuala%20Lumpur'
CITY_NAME2 = 'Paris'
CITY_NAME3 = 'Nairobi'

url_kul = f'https://api.openweathermap.org/data/2.5/forecast?q={CITY_NAME1}&appid={API_KEY}'
url_par = f'https://api.openweathermap.org/data/2.5/forecast?q={CITY_NAME2}&appid={API_KEY}'
url_nob = f'https://api.openweathermap.org/data/2.5/forecast?q={CITY_NAME3}&appid={API_KEY}'

response_kul = requests.get(url_kul)
response_par = requests.get(url_par)
response_nob = requests.get(url_nob)

response_kul = response_kul.json()
response_par = response_par.json()
response_nob = response_nob.json()
```

```python
print(response_kul['city']['name'])

ws_kul = []
temp_kul = []
hum_kul = []

for i in range(len(response_kul['list'])):
    ws_kul.append(response_kul['list'][i]['wind']['speed'])
    temp_kul.append(response_kul['list'][i]['main']['temp'])
    hum_kul.append(response_kul['list'][i]['main']['humidity'])
    
print(response_par['city']['name'])

ws_par = []
temp_par = []
hum_par = []

for i in range(len(response_par['list'])):
    ws_par.append(response_par['list'][i]['wind']['speed'])
    temp_par.append(response_par['list'][i]['main']['temp'])
    hum_par.append(response_par['list'][i]['main']['humidity'])
    
print(response_nob['city']['name'])

ws_nob = []
temp_nob = []
hum_nob = []

for i in range(len(response_nob['list'])):
    ws_nob.append(response_nob['list'][i]['wind']['speed'])
    temp_nob.append(response_nob['list'][i]['main']['temp'])
    hum_nob.append(response_nob['list'][i]['main']['humidity'])
```

```python
import matplotlib.pyplot as plt

plt.plot(ws_kul, 'g--',label = 'Kuala Lumpur')
plt.plot(ws_par, 'r--', label = 'Paris')
plt.plot(ws_nob, 'b--', label = 'Nairobi')
plt.ylabel('Wind Speed')
plt.legend()
plt.show()

plt.plot(temp_kul, 'g--',label = 'Kuala Lumpur')
plt.plot(temp_par, 'r--',label = 'Paris')
plt.plot(temp_nob, 'b--',label = 'Nairobi')
plt.ylabel('Current Temperature')
plt.legend()
plt.show()

plt.plot(hum_kul, 'g--',label = 'Kuala Lumpur')
plt.plot(hum_par, 'r--',label = 'Paris')
plt.plot(hum_nob, 'b--',label = 'Nairobi')
plt.ylabel('Humidity')
plt.legend()
plt.show()
```
