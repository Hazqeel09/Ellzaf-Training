```python
import matplotlib.pyplot as plt

data_file = open("dow_jones_index.data","r")

date_list = []

high_list = []
volume_list = []
pcp_list = []

avg_high_list = []
avg_volume_list = []
avg_pcp_list = []

next(data_file) #skip to next line
for data in data_file:
    data = data.split(',')
    date_list.append(data[2])
    #average high
    data[4] = float(data[4][1:])
    high_list.append(data[4])
    
    data[7] = int(data[7])
    volume_list.append(data[7])
    
    data[8] = float(data[8])
    pcp_list.append(data[8])

data_file.close()

for i in range(12):
    #print("------------------------")
    total_high = 0
    total_volume = 0
    total_pcp = 0
    for j in range(0,30*12,12): #based on data, there are 30 stocks and 12 weeks for each data for every quarter
        #print(date_list[j+i])
        total_high += high_list[j+i]
        total_volume += volume_list[j+i]
        total_pcp += pcp_list[j+i]
    
    avg_high_list.append(total_high/30)
    avg_volume_list.append(total_volume/30)
    avg_pcp_list.append(total_pcp/30)

    
# print(avg_high_list)
# print(avg_volume_list)
# print(avg_pcp_list)


plt.plot(avg_high_list, 'g--')
plt.ylabel('Average Highest Weekly Value')
plt.show()

plt.plot(avg_volume_list, 'r--')
plt.ylabel('Average Volume Weekly')
plt.show()

plt.plot(avg_pcp_list, 'b--')
plt.ylabel('Average Price Change Percent Weekly')
plt.show()
```
