```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin

x_range = []
for i in (np.linspace(-1, 1, 20001, True, 0.001)[0]):
    x_range.append(i)
    
xsine_function = [] #y = xsin(1/x)
for x in (np.linspace(-1, 1, 20001, True, 0.001)[0]): 
    if x == 0:
        xsine_function.append(0) #to avoid divide by 0
    else:
        xsine_function.append(x*(sin(1/x)))
        
    
plt.figure(figsize=(10,10))
plt.plot(x_range, xsine_function, 'm--')
plt.title('y = xsin(1/x)')
plt.show()
```

```python
from math import pi

x_range = []
for i in (np.linspace(0, 3, 3001, True, 0.001)[0]):
    x_range.append(i)

bouncy = [] #y = |Sin(x^x)/2^((x^x - pi/2)/pi)
for x in (np.linspace(0, 3, 3001, True, 0.001)[0]):
    bouncy.append(abs((sin(x ** x)) / (2 ** (((x ** x) - pi/2)/pi))))
    
plt.figure(figsize=(10,10))
plt.plot(x_range, bouncy, 'm--')
plt.title('y = |Sin(x^x)/2^((x^x - pi/2)/pi)')
plt.show()
```

```python
from math import factorial

x_range = []
for i in range(31):
    x_range.append(sin(i))
    
scribble = [] # sin y = sin(x!)
for x in range(31): 
    scribble.append(sin(factorial(x)))
    
plt.figure(figsize=(10,10))
plt.plot(x_range, scribble, 'm--')
plt.title('sin y = sin(x!)')
plt.show()
```
